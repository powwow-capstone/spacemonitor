```python
'''
Given a csv file with the following format:

OBJECTID,Crop2014,Acres,County,Comments,Source,Modified_B,Date_Data_,Last_Modif,DWR_Standa,GlobalID,Shape_Leng,Shape_Area,_count,_sum,_mean,_median,_stdev,_min,_max,_range,_minority,_majority,_variety,_variance

draw a histogram for every crop, x coordinate is the average eta with range between 1 and 8
y coordinate is the number of fields with eta value fall betweem 1 and 2, 2 and 3, etc for a specific crop. 
'''
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from sklearn.cluster import KMeans
```

```python
'''
create a dictionary with key representing crop name and value being a list of average eta values for that crop
from the given csv
'''
def create_crop_eta_dict(filename):
    crops = {}
    with open(filename, mode='r') as f:
        #skip the header row
        next(f, None)
        reader = csv.reader(f)
        
        for r in reader:
            crop = r[1]
            avg_eta = float(r[15])
            
            if crop not in crops:
                crops[crop] = []
            crops[crop].append(avg_eta)
    return crops
```

```python
crops_dict = create_crop_eta_dict("../data/CropCensus_withEta.csv")
```

```python
'''
function that draws a histogram given a list of eta values and crop name.
It also has an option to save the figure to a file
'''
def draw_hist_by_crop(eta_arr, crop, save_to_file = False):
    plt.title('average eta for ' + crop)
    plt.xlabel('average eta')
    plt.ylabel('number of fields')
    plt.hist(x=np.array(eta_arr, dtype=np.float64), bins=8)
    if save_to_file:
        plt.savefig('../plots/' + crop + '_eta.png')
    plt.show()
```

```python
draw_hist_by_crop(crops_dict["Grapes"], "Grapes", save_to_file = True)
```

```python
for key, value in crops_dict.items():
    draw_hist_by_crop(value, key, save_to_file = True)
```

```python
# compute the standard deviation for mean eta of every crop
for key, value in crops_dict.items():
    print(key)
    print(np.std(np.array(value, dtype=np.float64)))
```

```python
def fit_norm_by_crop(eta_arr, crop, save_to_file = False):
    mean, std=norm.fit(np.array(eta_arr, dtype=np.float64))
    plt.hist(np.array(eta_arr, dtype=np.float64), bins=8, density=True)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    y = norm.pdf(x, mean, std)
    plt.plot(x, y)
    if save_to_file:
        plt.savefig('../plots/' + crop + '_eta_normFit.png')
    plt.show()
```

```python
fit_norm_by_crop(crops_dict["Grapes"], "Grapes")
```

```python
'''
Given the csv file as described above. run a k-means clustering algorithm on every crop
cluster based on the x, y coordinates of the fields, group together fields that are close by for every crop
and calculate the eta mean and standard deviation for each cluster
'''
```

```python
'''
create a dictionary with key representing crop name and value being a list of lists [x,y,mean_eta] 
'''
def create_crop_xy_dict(filename):
    crops = {}
    with open(filename, mode='r') as f:
        #skip the header row
        next(f, None)
        reader = csv.reader(f)
        
        for r in reader:
            crop = r[1]
            
            x = float(r[25])
            y = float(r[26])
            avg_eta = float(r[15])
            
            if crop not in crops:
                crops[crop] = []
            crops[crop].append([x,y,avg_eta])
    return crops
```

```python
'''
Given a dictionary (key: crop, value: list of [x,y,avg_eta]), crop name, number of clusters
it runs the kmeans clutering algorithm for a given crop and returns a list if labels indicating the cluster id
and a list of [x,y,avg_eta]
'''
def cluster_by_crop(crops, crop, nclusters):
    xyeta = crops[crop]
    vec = np.zeros(shape=(len(xyeta),2))
    for i, v in enumerate(xyeta):
        vec[i] = v[:2]
    kmeans = KMeans(n_clusters=nclusters).fit(vec)
    return kmeans.labels_, xyeta
```

```python
'''
Run the kmeans clustering algorithm in a loop with k values from 1 to 10 and compute the SSE (Sum of Squared Error) for each k
The SSE is defined as the sum of the squared distance between each member of the cluster and its centroid.
plot the SSE for different k
based on the Elbow Criterion, we want to the pick a small k that still has a low SSE but it's a point where we start to have diminishing returns by increasing k.
'''
def cluster_by_crop_loop(crops, crop):
    xyeta = crops[crop]
    vec = np.zeros(shape=(len(xyeta),2))
    KMeans_labels_all = []
    sse = dict()
    for i, v in enumerate(xyeta):
        vec[i] = v[:2]
    for k in range(1, 10):
        if len(xyeta) < k:
            break
        kmeans = KMeans(n_clusters=k).fit(vec)
        KMeans_labels_all.append(kmeans.labels_)
        sse[k] = kmeans.inertia_
    plt.figure()
    plt.plot(list(sse.keys()), list(sse.values()))
    plt.title(crop)
    plt.xlabel("Number of clusters")
    plt.ylabel("SSE")
    plt.show()
    # return KMeans_labels_all, xyeta
```

```python
'''
Given list of labels returned from kmeans and the list of [x,y,eta]
it computes the mean and standard deviation for every cluster 
and returns a list of tupels (mean, std) for clusters related to a specific crop
'''
def compute_cluster_stats(labels, xyeta):
    cluster_eta = dict()
    cluster_stats = []
    for i,label in enumerate(labels):
        if label not in cluster_eta:
            cluster_eta[label] = []
        single_xyeta = xyeta[i]
        cluster_eta[label].append(single_xyeta[2])
    for cluster, eta in cluster_eta.items():
        mean = np.mean(eta)
        std = np.std(eta)
        cluster_stats.append((mean, std))
    return cluster_stats
```

```python
crop_xy_dict = create_crop_xy_dict("../data/CropCensus_withETa_and_XY_cleaned.csv")
```

```python
labels, xyetas = cluster_by_crop(crop_xy_dict, "Grapes", 5)
```

```python
'''
run the clustering all at once on all the crops with different k values
'''
def clustering(crop_xy_dict, nclusters, write_to_file = False):
    if write_to_file:
        with open("../data/Crop_xy_clusters.csv", 'w') as out_file:
            writer = csv.writer(out_file)
            writer.writerow(["crop", "number of clusters", "(mean, std)"])
            for crop, xy in crop_xy_dict.items():
                labels, xyetas = cluster_by_crop(crop_xy_dict, crop, nclusters[crop])
                cluster_stats = compute_cluster_stats(labels, xyetas)
                writer.writerow([crop, nclusters[crop], cluster_stats])
```

```python
cluster_by_crop_loop(crop_xy_dict, "Grapes")
```

```python
for crop, xy in crop_xy_dict.items():
    cluster_by_crop_loop(crop_xy_dict, crop)
```
